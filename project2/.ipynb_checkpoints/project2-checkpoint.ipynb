3 Columns have the delimter: cast, director, and genres. Let's take a sample and see some dataset and if really is a delimter in the columns other than cast:

```python
# Look for an index where director has the delimter:
df_clean[df_clean.director.str.contains('\|')].director.sample()
```

```python
# Look for an index where director has the delimter:
df_clean[df_clean.genres.str.contains('\|')].genres.sample()
```

Yup, as expected. Delimter in those columns does seperate different values (a number of directors of a movie, or different genres of a movie. To process them properly, let's see the unique number of values for each cast, director, and genre.

```python
def unique_values(series):
    s = set()
    for row in series:
        s.update(row.split('|'))
    return s

print("There are: {} unique director".format(len(unique_values(df_clean.director))))
print("There are: {} unique genre".format(len(unique_values(df_clean.genres))))
print("There are: {} unique cast".format(len(unique_values(df_clean.cast))))
```

There are multiple ways to organize these details into the dataframe, a couple I could think of:

1) Hot encoding (dummies): which will add all unique values are columns. In case of the cast, my dataframe will have an additional 18930 columns. Wow, that's a lot to process and my device cannot handle that.

2) extend my dataframe to include 

So, for genres, I'll just go with hot encoding, while extending both casts and directors

```python
# Genres Hot encoding 
genres_dummies = df_clean.genres.str.get_dummies('|')
df_clean = df_clean.merge(genres_dummies, left_on='id', right_on='id')
list_of_genres = genres_dummies.columns
```
